```python
tips= sns.load_dataset("tips")
```

```python
tips
```

```python
sns.scatterplot(x=tips.total_bill , y = tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x =tips.total_bill , y = tips.tip , data = tips, hue='smoker' ) #hue - color coding method to differentiate
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x =tips.total_bill , y = tips.tip , data = tips, style = 'size' )
```

```python
sns.relplot(x =tips.total_bill , y = tips.tip , data = tips, hue='time' , style= "size") #to check duration at which time people arrived to have food
```

```python
sns.catplot(x ='day' , y = 'total_bill' , data = tips)
```

```python
sns.jointplot(x= tips.total_bill , y = tips.tip)
```

```python
sns.pairplot(tips)
```
